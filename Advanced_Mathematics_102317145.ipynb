```python
import numpy as np
```

**Step 1:Transformation of x into z**

```python
r=102317145
ar=0.05*(r%7)
br=0.3*((r%5)+1)
print(f"Using roll number: {r}")
print(f"Computed coefficient ar = {ar}")
print(f"Computed coefficient br = {br}")
```

```python

z=x+ar*np.sin(br*x)
print("Transformed values:",z)
```

**Step 2:Learning parameters of the given probability density function using Method of
 Moments**

```python
mean1 = np.mean(z)
variance = np.var(z)
lambda1 = 1 / (2 * variance)
c = np.sqrt(lambda1 / np.pi)
```

```python
print("Estimated PDF Parameters:")
print("value of lambda:",lambda1)
print("value of mu:",mean1)
print("value of c:",c)
```
